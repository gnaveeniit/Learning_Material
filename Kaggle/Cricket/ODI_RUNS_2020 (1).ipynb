```
#lets look at the Pie chart of the percentage of RUNS scored by each country

runs_AUS=ODI_runs.loc[ODI_runs['Country']=='AUS']['Runs'].sum()
runs_NZ=ODI_runs.loc[ODI_runs['Country']=='NZ']['Runs'].sum()
runs_INDIA=ODI_runs.loc[ODI_runs['Country']=='INDIA']['Runs'].sum()
runs_PAK=ODI_runs.loc[ODI_runs['Country']=='PAK']['Runs'].sum()
runs_SL=ODI_runs.loc[ODI_runs['Country']=='SL']['Runs'].sum()
runs_BNG=ODI_runs.loc[ODI_runs['Country']=='BNG']['Runs'].sum()
runs_WI=ODI_runs.loc[ODI_runs['Country']=='WI']['Runs'].sum()
runs_ENG=ODI_runs.loc[ODI_runs['Country']=='ENG']['Runs'].sum()
runs_ZIM=ODI_runs.loc[ODI_runs['Country']=='ZIM']['Runs'].sum()
runs_SA=ODI_runs.loc[ODI_runs['Country']=='SA']['Runs'].sum()

runs=[runs_AUS,runs_NZ,runs_INDIA,runs_PAK,runs_SL,runs_BNG,runs_WI,runs_ENG,runs_ZIM,runs_SA]
labels=['AUS','NZ','INDIA','PAK','SL','BNG','WI','ENG','ZIM','SA']

plt.style.use('ggplot')
plt.style.use('default')
explode = (0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05)
plt.figure(figsize=(10,8),dpi=100)
plt.title('ICC: ODI Runs scored by Country by Top 90 players',fontdict={'fontweight':'bold', 'fontsize': 12})
#plt.legend()

plt.pie(runs, labels = labels, explode=explode,counterclock=True,shadow=False,labeldistance=1.1,
        pctdistance=0.85, autopct='%.2f %%') 

plt.show()

#Stats:Its suprising that England has only 3.47% by its players in top 90 ODI runs scorers
```

```
# lets look at the Highest Scored by these top 90 players using SCATTER PLOTb w.r.t to number of fifties
HS_Country=ODI_runs[['Player','Country','HS','4s',50]]
HS_Country

#We have a problem of '*' in HS
```

```
# Firstly, we have to make sure to remove '*' in HS
HS_Country.HS = [int(x.strip('*')) if type(x)==str else x for x in HS_Country['HS']]
HS_Country
#Now that we removed the '*' from HS, lets mover further
```

```
#Plotting scatter plot

#plt.style.use('seaborn')
plt.figure(figsize=(7,5),dpi=105)
x=HS_Country[50]
y=HS_Country.HS
#groups = '', "", "")
teams=['AUS','NZ','INDIA','PAK','SL','BNG','WI','ENG','ZIM','SA']

plt.scatter(x, y,alpha=0.8,c ="b",s=20,edgecolors='red',linewidth=1)

plt.title('Scatter plot between HS & number of fifties',fontdict={'fontweight':'bold', 'fontsize': 18})
plt.legend(loc=2)
plt.xlabel('Number of fifties in a career')
plt.ylabel('HS in their career')

plt.show()
```
